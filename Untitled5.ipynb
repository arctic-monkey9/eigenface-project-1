```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

face_data = mat_content['X']

label_data = mat_content['l']

train_set = []
test_set = []
train_label = []
test_label = []

for i in range(0, 520):
    if i % 5 == 0:
        test_set.append(face_data[:, i])
        test_label.append(label_data[:, i])
    else:
        train_set.append(face_data[:, i])
        train_label.append(label_data[:, i])

# 훈련 세트에서 평균 얼굴과 주성분(PCA 기저)을 계산합니다.
pca = PCA()
pca.fit(train_set)

# 주성분의 수를 변경하여 재구성을 수행합니다.
num_bases_list = [50,60,70, 80]  # 사용할 기저의 수 예시
for num_bases in num_bases_list:
    pca.n_components = num_bases
    pca.fit(train_set)
    reconstructed_images = pca.inverse_transform(pca.transform(test_set))

    # 재구성 오차/왜곡을 측정합니다.
    reconstruction_errors = np.mean(np.square(np.array(test_set) - reconstructed_images), axis=1)

    # 선택한 이미지에 대한 결과를 표시하고 논의합니다.
    for i in range(3):  # 3개 이미지에 대한 결과 표시
        original_image = np.reshape(test_set[i], (46,56)).T
        reconstructed_image = np.reshape(reconstructed_images[i], (46,56)).T

        # 원본 이미지와 재구성된 이미지를 표시합니다.
        plt.figure(figsize=(8, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(original_image, cmap='gray')
        plt.title('원본 이미지')
        plt.subplot(1, 2, 2)
        plt.imshow(reconstructed_image, cmap='gray')
        plt.title('재구성된 이미지')
        plt.show()

        # 이론적 값과 재구성 오차를 비교합니다.
        print(f"{num_bases}개 기저를 사용한 {i + 1}번째 이미지의 재구성 오차: {reconstruction_errors[i]}")

```

```python

from pandas._libs.tslibs import timestamps
from google.colab import drive
drive.mount('/content/drive')

import scipy.io as sio
import numpy as np
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import time

# 데이터 불러오기
mat_content = sio.loadmat("/content/drive/MyDrive/face.mat")
face_data = mat_content['X']
label_data = mat_content['l']

# 데이터 중심화 (centering)
mean_face = np.mean(face_data, axis=1, keepdims=True)
centered_face_data = face_data - mean_face

# 데이터 분할
train_set1 = centered_face_data[:, ::5]
train_set4 = centered_face_data[:, 1::5]
train_set3 = centered_face_data[:, 2::5]
train_set2 = centered_face_data[:, 3::5]
test_set = centered_face_data[:, 4::5]

train_label_set1 = label_data[:, ::5]
train_label_set4 = label_data[:, 1::5]
train_label_set3 = label_data[:, 2::5]
train_label_set2 = label_data[:, 3::5]
test_label_set = label_data[:, 4::5]

n_components = 63  # PCA의 주성분 수 설정
n_neighbors = 5
# PCA for train_set1
pca_train_set1 = PCA(n_components=n_components)
start_time = time.time()
pca_train_set1.fit(train_set1.T)
training_time_train_set1 = time.time() - start_time

# Reconstruction Error 계산
reconstructed_train_set1 = pca_train_set1.inverse_transform(pca_train_set1.transform(train_set1.T))
reconstruction_error_train_set1 = mean_squared_error(train_set1.T, reconstructed_train_set1)

# NN Classification을 사용한 Face Recognition Accuracy 계산
knn_train_set1 = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_train_set1.fit(pca_train_set1.transform(train_set1.T), np.ravel(train_label_set1))
test_set_pca = pca_train_set1.transform(test_set.T)
predictions_train_set1 = knn_train_set1.predict(test_set_pca)
accuracy_train_set1 = accuracy_score(np.ravel(test_label_set), predictions_train_set1)

from sklearn.decomposition import IncrementalPCA

train_sets = [train_set1, train_set2, train_set3, train_set4]
train_label_sets = [train_label_set1, train_label_set2, train_label_set3, train_label_set4]

training_times_incremental = []
reconstruction_errors_incremental = []
accuracies_incremental = []

combined_train_set = np.concatenate(train_sets[:4], axis=1).T
combined_train_labels = np.concatenate(train_label_sets[:4], axis=1).T

pca_incremental = IncrementalPCA(n_components = n_components, batch_size = 104)
start_time = time.time()
pca_incremental.fit(combined_train_set)

training_time = time.time() - start_time
training_times_incremental.append(training_time)

reconstructed_combined = pca_incremental.inverse_transform(pca_incremental.transform(combined_train_set))
reconstruction_error = mean_squared_error(combined_train_set, reconstructed_combined)
reconstruction_errors_incremental.append(reconstruction_error)
knn_incremental = KNeighborsClassifier(n_neighbors=n_neighbors)
print(pca_incremental.transform(combined_train_set).shape)
knn_incremental.fit(pca_incremental.transform(combined_train_set), np.ravel(combined_train_labels))

test_set_pca = pca_incremental.transform(test_set.T)
predictions_incremental = knn_incremental.predict(test_set_pca)
accuracy_incremental = accuracy_score(np.ravel(test_label_set), predictions_incremental)
accuracies_incremental.append(accuracy_incremental)

# 전체 데이터에 대한 PCA
combined_train_set = np.concatenate(train_sets[:4], axis=1).T
combined_train_labels = np.concatenate(train_label_sets[:4], axis=1).T

pca_full = PCA(n_components=n_components)
start_time = time.time()
pca_full.fit(combined_train_set)
training_time_full = time.time() - start_time

reconstructed_full = pca_full.inverse_transform(pca_full.transform(combined_train_set))
reconstruction_error_full = mean_squared_error(combined_train_set, reconstructed_full)

knn_full = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_full.fit(pca_full.transform(combined_train_set), np.ravel(combined_train_labels))
test_set_pca = pca_full.transform(test_set.T)
predictions_full = knn_full.predict(test_set_pca)
accuracy_full = accuracy_score(np.ravel(test_label_set), predictions_full)

# 결과 출력
print("PCA의 주성분 수: ", n_components)
print("NN classification의 neighbor 수:", n_neighbors)
print("PCA for train_set1:")
print("Training Time:", training_time_train_set1)
print("Reconstruction Error:", reconstruction_error_train_set1)
print("Accuracy:", accuracy_train_set1)
print()

print("Incremental PCA:")
print("Training Time:", training_times_incremental[0])
print("Reconstruction Error:", reconstruction_errors_incremental[0])
print("Accuracy:", accuracies_incremental[0])
print()

print("PCA for Full Data:")
print("Training Time:", training_time_full)
print("Reconstruction Error:", reconstruction_error_full)
print("Accuracy:", accuracy_full)










```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

# Add noise to feature space for randomization
def add_noise(data, noise_factor=0.1):
    noisy_data = data + noise_factor * np.random.normal(loc=0, scale=1, size=data.shape)
    return noisy_data

# Bagging for sample randomization
def apply_bagging(data, labels, num_bagging_samples=5):
    bagged_data = []
    bagged_labels = []
    for i in range(num_bagging_samples):
        bag_idx = np.random.choice(len(data), len(data), replace=True)
        bagged_data.append(data[bag_idx])
        bagged_labels.append(labels[bag_idx])

    return np.array(bagged_data), np.array(bagged_labels)

face_data = mat_content['X']
label_data = mat_content['l']
test_set = []
test_labels = []
train_set = []
train_labels = []

for i in range(0, 520):
    if i % 5 == 3:
        test_set.append(face_data[:, i])
        test_labels.append(label_data[:, i])
    else:
        train_set.append(face_data[:, i])
        train_labels.append(label_data[:, i])

M_pca_values = [50, 100, 150]
M_lda_values = [10, 20, 30]
num_bagging_samples = 5

for M_pca in M_pca_values:
    for M_lda in M_lda_values:
        pca = PCA(n_components=M_pca)
        X_train_pca = pca.fit_transform(train_set)
        X_test_pca = pca.transform(test_set)

        lda = LinearDiscriminantAnalysis(n_components=min(M_lda, len(np.unique(label_data)) - 1))
        X_train_lda = lda.fit_transform(X_train_pca, np.ravel(train_labels))
        X_test_lda = lda.transform(X_test_pca)

        # Adding noise to feature space
        X_train_lda_noisy = add_noise(X_train_lda)
        X_test_lda_noisy = add_noise(X_test_lda)

        # Applying bagging for sample randomization
        X_train_bagged, train_labels_bagged = apply_bagging(X_train_lda_noisy, np.array(train_labels), num_bagging_samples)

        base_models = []
        for i in range(num_bagging_samples):
            nn_classifier = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=300)
            nn_classifier.fit(X_train_bagged[i], np.ravel(train_labels_bagged[i]))
            base_models.append(nn_classifier)

        committee_machine_errors = []
        individual_model_errors = []
        for i in range(num_bagging_samples):
            committee_machine_errors.append(np.mean(np.array([base_models[j].predict(X_test_lda_noisy) for j in range(i + 1)]), axis=0))
            individual_model_errors.append(base_models[i].predict(X_test_lda_noisy))

        fusion_rules = ["mean", "max", "voting"]  # Different fusion rules to compare

        for rule in fusion_rules:
            if rule == "mean":
                committee_machine_error = np.mean(committee_machine_errors, axis=0)
            elif rule == "max":
                committee_machine_error = np.max(committee_machine_errors, axis=0)
            elif rule == "voting":
                committee_machine_error = np.round(np.mean(committee_machine_errors, axis=0))

            print(f"Fusion rule: {rule}")
            print(f"Committee machine error: {np.mean(committee_machine_error != np.ravel(test_labels))}")
            print(f"Average individual model error: {np.mean([np.mean(individual_model_errors[i] != np.ravel(test_labels)) for i in range(num_bagging_samples)])}")

            y_pred = nn_classifier.predict(X_test_lda)
            conf_matrix = confusion_matrix(np.ravel(test_labels), y_pred)
            print("Confusion matrix:")
            print(conf_matrix)














```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import time
import random

# 데이터 준비
face_data = mat_content['X']
label_data = mat_content['l']

train_set = []
test_set = []
train_labels = []
test_labels = []

for i in range(0, 520):
    if i % 5 == 3:
        test_set.append(face_data[:, i])
        test_labels.append(label_data[:, i])
    else:
        train_set.append(face_data[:, i])
        train_labels.append(label_data[:, i])

# 랜덤 포레스트 모델 훈련 및 테스트
n_estimators_values = [10, 50, 100]  # 트리의 개수를 변경해보세요
max_depth_values = [None, 10, 20]  # 트리의 최대 깊이를 변경해보세요
max_features_values = ['auto', 'sqrt', 'log2']  # 최대 특징 수를 변경해보세요

for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        for max_features in max_features_values:
            start_time = time.time()
            rf_classifier = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, max_features=max_features)
            rf_classifier.fit(train_set, train_labels)
            end_time = time.time()

            # 정확도 및 혼동 행렬 측정
            predicted_labels = rf_classifier.predict(test_set)
            accuracy = accuracy_score(test_labels, predicted_labels)
            conf_matrix = confusion_matrix(test_labels, predicted_labels)

            # 결과 출력
            print(f"Number of estimators: {n_estimators}, Max depth: {max_depth}, Max features: {max_features}")
            print(f"Accuracy: {accuracy}")
            print("Confusion matrix:")
            print(conf_matrix)
            print(f"Time taken: {end_time - start_time} seconds")
            print("------------")

            # 랜덤 예제 선택 및 결과 시각화
            random_example_idx = random.randint(0, len(test_set) - 1)
            example_image = test_set[random_example_idx]
            predicted_label = rf_classifier.predict([example_image])[0]
            true_label = test_labels[random_example_idx]

            if predicted_label == true_label:
                print("Example Success:")
            else:
                print("Example Failure:")

            print(f"True Label: {true_label}, Predicted Label: {predicted_label}")

            # 랜덤한 예제 이미지 플로팅
            plt.imshow(np.reshape(example_image, (46, 56)).T, cmap='gray')
            plt.show()

```
